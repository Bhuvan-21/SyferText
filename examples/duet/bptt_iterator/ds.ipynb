```python
import syft as sy
# duet = sy.join_duet("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
duet = sy.join_duet(loopback=True)
```

```python
sy.load_lib('syfertext')
```

Load SyferText support in Syft

```python
import syfertext
```

```python
from syfertext.tokenizers import SpacyTokenizer
from syfertext.vocab import Vocab
from syfertext.data.readers.language_modeling import TextReader
from syfertext.data.iterators.language_modeling import BPTTIterator

from syfertext.encoders import SentenceEncoder
```

```python
# Create tokenizer
tokenizer = SpacyTokenizer(exceptions={})

# Create a vocab
vocab = Vocab()

# Create an encoder
encoder = SentenceEncoder(tokenizer = tokenizer,
                          vocab = vocab
                         )
```

Create a reader

```python
lm_reader = TextReader(encoder = encoder,
                        mode = 'train')
```

Test the bptt iterator

```python
bptt_iterator = BPTTIterator(batch_size = 4,
                             bptt_len = 6,
                             dataset_reader = lm_reader
                            )
```

```python
bptt_iterator_ptr = bptt_iterator.send(duet, searchable = True, 
                                       tags = ['#bptt_iterator'], 
                                       description = "I am a SyferText bptt iterator")
```
